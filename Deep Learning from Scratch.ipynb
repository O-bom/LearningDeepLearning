## Python script

```python
# class 만들기
class Man:
    def __init__(self, name):
        self.name = name
        print("Initialized!")

    def hello(self):
        print("Hello " + self.name + "!")

    def goodbye(self):
        print("Good-bye " + self.name + "!")

m = Man("David")
m.hello()
m.goodbye()
```

```python
# 2) ndarray의 초기화
# 모든 값이 0인 행렬 생성 : np.zeros()
zero_mat = np.zeros((2,3))
print(zero_mat)
```

```python
# 모든 값이 1인 행렬 생성 : np.ones()
one_mat = np.ones((2,3))
print(one_mat)
```

```python
# 모든 값이 특정 상수인 행렬 생성 : np.full()
same_value_mat = np.full((2,2),7)
print(same_value_mat)
```

```python
# 대각선 값이 1이고 나머지가 0인 행렬 생성 : np.eye
eye_mat = np.eye(3)
print(eye_mat)
```

```python
# 임의의 값으로 채워진 행렬 생성 : np.random.random()
random_mat = np.random.random((2,2))
print(random_mat)
```

```python
# 3) np.arange
# 0부터 n-1까지의 값을 가지는 행렬 생성 : np.arange(n)
range_vec1 = np.arange(10)
print(range_vec1)

# n부터 m-1까지의 값을 o만큼의 간격으로 행렬 생성 : np.arange(n,m,o)
range_vec2 = np.arange(1,10,2)
print(range_vec2)
```

```python
# 4) np.reshape
# 내부 데이터 변경하지 않으면서 배열의 구조 변경 : np.reshape(행,열))
reshape_mat = np.array(np.arange(1,31).reshape((3,10)))
print(reshape_mat)
```
